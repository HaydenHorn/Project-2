```python
X = dataset[[' Fixed Assets Turnover Frequency', ' Borrowing dependency',' Equity to Long-term Liability', ' Equity to Liability',
' Current Liabilities/Liability',' Total expense/Assets',' No-credit Interval',' Operating Gross Margin',
' Cash Flow to Liability', ' After-tax Net Profit Growth Rate', ' Cash Flow Per Share',' CFO to Assets',
' Gross Profit to Sales',' ROA(C) before interest and depreciation before interest',' Operating profit per person']]
y = dataset[['Bankrupt?']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print('r2_score',r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error as msc
from math import sqrt
rms=sqrt(msc(y_test,y_pred))
print('The mean square error is', rms)
```

```python
X = dataset[[' Fixed Assets Turnover Frequency', ' Borrowing dependency',' Equity to Long-term Liability', ' Equity to Liability',
' Current Liabilities/Liability',' Total expense/Assets',' No-credit Interval',' Operating Gross Margin',
' Cash Flow to Liability', ' After-tax Net Profit Growth Rate', ' Cash Flow Per Share',' CFO to Assets',
' Gross Profit to Sales',' ROA(C) before interest and depreciation before interest',' Operating profit per person']]
y = dataset[['Bankrupt?']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler()
sc_y=StandardScaler()
X_train=sc_X.fit_transform(X_train)
y_train=sc_y.fit_transform(y_train)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(X_train,y_train)
```

```python
y_pred=sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test)).reshape(-1,1))
y_pred
```

```python
from sklearn.metrics import r2_score
print('r2_score',r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error as msc
from math import sqrt
rms=sqrt(msc(y_test,y_pred))
print('The mean square error is', rms)
```
